```python
ax=sns.countplot(x="Size",data=df) # x eksenin size bedenleri sütünü yer alır ax=axes objesini gösterir grafigin nerede çizilicegini belirler.
# birden fazla grafigi aynı eksende gösterir. data=df seaborn hangi verileri gösterecegini bilmesi için eşitlenir.
```

```python
ax=sns.countplot(x='Size' ,data=df) # countpolot seabornde sayım grafiğini oluşturur.grafiklerin üzerinde 5937 4504 5516 gibi
for bars in ax.containers: #ax.containers   grafikteki çubukları temsil eder Her çubuk için bir "bars" nesnesi döndürür. for bars in eklenince döndüdür hepsini
    ax.bar_label(bars) # belirli bir çubuk grubunun üstüne değer etiketleri eklemek için kullanılır. M XL L gibi
# Grafikten insanların çoğunun M-Boyut satın aldığını görebilirsiniz
```

```python
df.groupby(['Size'], as_index=False)['Qty'].sum().sort_values(by='Qty',ascending=False) # as_index=False  Bu parametre, grup anahtarlarının (burada "Size") bir indeks olarak değil,
# normal bir sütun olarak döndürülmesini sağlar. normal sutun görünmesi için False kullanıldı
```

```python


# Beden bazında  gruplama ve toplamını hesaplama
top_selling_size = df.groupby('Size')['Qty'].sum().idxmax()

# En faza  beden miktarı bulma
print(f"En fazla satılan  beden miktari: {top_selling_size}")
```

```python
S_Qty=df.groupby(['Size'], as_index=False)['Qty'].sum().sort_values(by='Qty',ascending=False) #sort_values sıralama işlemleridir ascending artanmı azalanmı diye gösterir
#Ascending de True artan False azalan olur
sns.barplot(x='Size',y='Qty', data=S_Qty)
plt.show()
#  Grafikte satışlarda Miktarın çoğunun M-Boyut satın aldığını görebilirsiniz.
```

```python
plt.figure(figsize=(10,5)) #figür boyutunu ayarlar figsize
ax=sns.countplot(data=df, x='Courier Status',hue= 'Status') # hue='Status'  çubukların farklı renklerde gösterilmesini sağlar.
plt.show()
#  Grafikte siparişlerin çoğunluğu kuryeler aracılığıyla gönderilmektedir.
```

```python

df['Size'].hist() #veri dagılımını gösteren histogramdır.
plt.show()
```

```python
df['Category'] = df['Category'].astype(str)
column_data = df['Category']
plt.figure(figsize=(10, 5)) #genişlik 10inç yükseklik 5 inç dir.
plt.hist(column_data, bins=30, edgecolor='Black') #edgecolor histogram cubuklarının rengini siyah yapar bins histogram çubuğunu 30 aralıga kadar böler çubukları küçültür büyültür.
plt.xticks(rotation=0) # rotation xticks etiketlerin yazısını hizalar kurta set bottom gibi
plt.show()
# Grafikte alıcıların çoğunun Set olduğunu görebilirsiniz.
```

```python
# # Pasta grafiğini kullanarak B2B Verilerini Kontrol Etme
B2B_Check = df['B2B'].value_counts() #benzersiz degerleri gösterir
#  Pasta grafiğini çizer
plt.pie(B2B_Check, labels=B2B_Check, autopct='%1.1f%%')
plt.axis('equal')
plt.show()
# Grafikten görebileceğimiz gibi alıcıların maksimum %99,2'si perakendeci, %0,8'i ise B2B alıcısıdır.
```

```python

# Pasta grafiği için verileri hazırlamak
a1 = df['Fulfilment'].value_counts()
# Step 4: Plot the pie chart
fig, ax = plt.subplots()
ax.pie(a1, labels=a1.index, autopct='%1.1f%%', radius=0.7, wedgeprops=dict(width=0.6))#wedgeprops=dict(width=0.6) Pasta dilimlerinin kalınlığını ayarlamak için kullanılır dilimlerin kalınlığı 0.6 birim olarak ayarlanmıştır.
ax.set(aspect="equal") #radius=0.7 Pasta grafiğinin genel yarıçapını ayarlar. 0 ile 1 arasında bir değer alır; 1, tam daireyi temsil eder.
plt.show()
# Tabloda, Lojistik hizmetlerinin çoğunun Amazon tarafından sağlandığını görebilirsiniz.
# autopct='%1.1f%%': Pasta dilimlerinin üzerindeki yüzdeleri formatlamak için kullanılır. %1.1f%%, yüzdelerin bir ondalık basamakla gösterileceği anlamına gelir
#labels=a1.index: Pasta dilimlerinin etiketlerini ayarlamak için kullanılır.
```

```python
# Dağılım grafiği için verileri hazırlanır
x_data = df['Category']
y_data = df['Size']

# Dağılım grafiğini çizilir
plt.scatter(x_data, y_data)
plt.xlabel('Category')
plt.ylabel('Size')
plt.title('Scatter Plot')

# Daha iyi okunabilirlik için x ekseni etiketlerini döndürürüz
plt.xticks(rotation=45)

# Grafiği göstermek
plt.show()
```

```python
# Şehirlerin Arsa Sayısı
plt.figure(figsize=(12, 6))
sns.countplot(data=df, x='ship-state')
plt.xlabel('ship-state')
plt.ylabel('count')
plt.title('Distribution of State') #Devletin Dagılımı
plt.xticks(rotation=90)
plt.show()
```

```python
# İlk 10 Eyalet
top_10_state = df['ship-state'].value_counts().head(10)
# Eyaletlere göre şehirlerin arsa sayısı
plt.figure(figsize=(12, 6))
sns.countplot(data=df[df['ship-state'].isin(top_10_state.index)], x='ship-state')
plt.xlabel('ship-state')
plt.ylabel('count') # Miktar
plt.title('Distribution of  State')
plt.xticks(rotation=45)
plt.show()
#   Grafikten alıcıların çoğunun Maharashtra eyaletinden olduğunu görebilirsiniz.
```
